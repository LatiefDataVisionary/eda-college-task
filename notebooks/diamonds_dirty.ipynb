```python
import pandas as pd
import numpy as np
from google.colab import files

# Muat dataset asli
# uploaded = files.upload()
```

```python
df = pd.read_csv('diamonds.csv')
```

```python
df.head()
```

```python
# 1. Tambahkan Missing Values
np.random.seed(42)
missing_mask = np.random.choice([True, False], size=df.shape, p=[0.1, 0.9])  # 10% nilai menjadi NaN
df_missing = df.mask(missing_mask)
```

```python
# 2. Tambahkan Outlier
outlier_indices = np.random.choice(df.index, size=50, replace=False)
df_missing.loc[outlier_indices, 'price'] = df_missing['price'].max() * 10
```

```python
# 3. Kesalahan Data
invalid_indices_cut = np.random.choice(df.index, size=20, replace=False)
df_missing.loc[invalid_indices_cut, 'cut'] = 'InvalidCut'

invalid_indices_color = np.random.choice(df.index, size=20, replace=False)
df_missing.loc[invalid_indices_color, 'color'] = 'InvalidColor'
```

```python
# 4. Duplicate Value
# Menggandakan beberapa baris secara acak
duplicate_indices = np.random.choice(df.index, size=30, replace=False)
df_duplicate = df.loc[duplicate_indices]
df_with_duplicates = pd.concat([df, df_duplicate], ignore_index=True)
df = df_with_duplicates
```

```python
output_file = "diamonds_dirty.csv"
df.to_csv(output_file, index=False)

# Unduh file langsung di Google Colab
files.download(output_file)
```
