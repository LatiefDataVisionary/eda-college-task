**Visualisasi Data**

```python
plt.hist(df_gubernur['Perolehan Suara'])
plt.title('Distribusi Perolehan Suara Gubernur')
plt.xlabel('Perolehan Suara')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.hist(df_gubernur['Total Suara'])
plt.title('Distribusi Total Suara Gubernur')
plt.xlabel('Total Suara')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.scatter(df_gubernur['Perolehan Suara'], df_gubernur['Total Suara'])
plt.title('Hubungan Perolehan Suara dan Total Suara Gubernur')
plt.xlabel('Perolehan Suara')
plt.ylabel('Total Suara')
plt.show()
```
