**EDA dan Visualisasi Data**

```python
plt.hist(df_gubernur['Perolehan Suara'])
plt.title('Distribusi Perolehan Suara Gubernur')
plt.xlabel('Perolehan Suara')
plt.ylabel('Frekuensi')
plt.show()
```

Grafik histogram digunakan untuk melihat distribusi data:

Perolehan Suara Gubernur: Menunjukkan distribusi jumlah suara yang diperoleh kandidat.

Total Suara Gubernur: Menunjukkan distribusi total suara yang dihitung dalam pemilihan gubernur.

Kesimpulan Grafik Histogram:
Distribusi Perolehan Suara Gubernur: Jika histogram berbentuk skewed ke kanan, ini menunjukkan bahwa sebagian besar kandidat mendapatkan suara rendah, sementara beberapa mendapatkan suara sangat tinggi.

Distribusi Total Suara Gubernur: Distribusi total suara biasanya lebih merata, tetapi outlier dapat menandakan daerah dengan jumlah pemilih yang tidak proporsional.

```python
plt.hist(df_gubernur['Total Suara'])
plt.title('Distribusi Total Suara Gubernur')
plt.xlabel('Total Suara')
plt.ylabel('Frekuensi')
plt.show()
```

Correlation Analysis

```python
plt.scatter(df_gubernur['Perolehan Suara'], df_gubernur['Total Suara'])
plt.title('Hubungan Perolehan Suara dan Total Suara Gubernur')
plt.xlabel('Perolehan Suara')
plt.ylabel('Total Suara')
plt.show()
```

```python
plt.scatter(df_bupati_walkot['Perolehan Suara'], df_bupati_walkot['Total Suara'])
plt.title('Hubungan Perolehan Suara dan Total Suara Bupati dan Walikota')
plt.xlabel('Perolehan Suara')
plt.ylabel('Total Suara')
plt.show()
```

Scatter plot menggambarkan hubungan antara perolehan suara dan total suara:

Gubernur: Hubungan kandidat dengan total suara di wilayah masing-masing.

Bupati/Walikota: Sama seperti gubernur, tetapi fokus pada daerah tingkat II.

Kesimpulan Scatter Plot:
Jika terdapat korelasi positif, kandidat yang memperoleh suara lebih tinggi biasanya di wilayah dengan total suara tinggi (misalnya, wilayah padat penduduk).

Jika hubungan lemah atau tidak ada pola, faktor lain (seperti politik atau kandidat independen) mungkin berpengaruh.

```python
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation calculation
numeric_df = df_gubernur.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Gubernur')
plt.show()
```

Heatmap digunakan untuk menganalisis hubungan antar variabel numerik (misalnya, perolehan suara dan total suara):

Korelasi dekat ke +1: Hubungan positif kuat.

Korelasi dekat ke -1: Hubungan negatif kuat.

Korelasi dekat ke 0: Tidak ada hubungan.

Kesimpulan Heatmap:
Korelasi tinggi menunjukkan bahwa variabel saling memengaruhi, misalnya perolehan suara biasanya berbanding lurus dengan total suara.

Korelasi lemah mungkin menunjukkan ketidaksesuaian atau anomali di data.

```python
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation calculation
numeric_df = df_bupati_walkot.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix - Bupati dan Walikota')
plt.show()
```

Outlier Detection (using boxplots)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_gubernur[['Perolehan Suara', 'Total Suara']])
plt.title('Boxplot for Outlier Detection - Gubernur')
plt.show()
```

Boxplot digunakan untuk mendeteksi outlier dalam data:

Gubernur: Mendeteksi kandidat dengan perolehan suara jauh di atas atau di bawah rata-rata.

Bupati/Walikota: Sama seperti gubernur.

Kesimpulan Boxplot:
Outlier menunjukkan daerah atau kandidat yang luar biasa baik dari segi perolehan suara atau total suara.

Outlier dapat dianalisis lebih lanjut untuk mengetahui penyebabnya (misalnya, kandidat populer atau anomali data).

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_bupati_walkot[['Perolehan Suara', 'Total Suara']])
plt.title('Boxplot for Outlier Detection - Bupati dan Walikota')
plt.show()
```
